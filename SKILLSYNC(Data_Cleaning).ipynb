```python
import pandas as pd

df = pd.read_json('/content/resumes_dataset.jsonl', lines=True)

print(df.head())
```

```python
print(df.isnull().sum())
```

```python
df.dropna(inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
print(df.columns)
```

```python
import re

def clean_text(text):

    text = str(text)

    text = re.sub(r'http\S+', ' ', text)

    text = re.sub(r'www\S+', ' ', text)

    text = re.sub(r'[^A-Za-z0-9 ]+', ' ', text)

    text = re.sub(r'\s+', ' ', text)

    text = text.lower()

    return text
```

```python
df['Text'] = df['Text'].apply(clean_text)
```

```python
print(df['Text'].head())
```

```python
df.to_csv('clean_resume_dataset.csv', index=False)
```
