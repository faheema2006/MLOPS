```python
import pandas as pd
dataset = pd.read_csv('/content/clean_resume_dataset.csv')
print(dataset.head())
```

```python
inputs = dataset.drop('Category',axis='columns')
target = dataset['Category']
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=100)
```

```python
X_text = tfidf.fit_transform(inputs['Text']).toarray()
```

```python
category_encoder = LabelEncoder()
target_n = category_encoder.fit_transform(target)
```

```python
print(target_n)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_text, target_n)
```

```python
model.score(X_text, target_n)
```

```python
sample_resume = ["""Python SQL Machine Learning Data Analysis Deep Learning TensorFlow NLP"""]
```

```python
sample_vector = tfidf.transform(sample_resume).toarray()
```

```python
prediction = model.predict(sample_vector)
print(prediction)
```

```python
predicted_category = category_encoder.inverse_transform(prediction)
print(predicted_category)
```
