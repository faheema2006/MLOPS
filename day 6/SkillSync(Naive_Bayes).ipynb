```python
import pandas as pd
dataset = pd.read_csv('/content/clean_resume_dataset.csv')
print(dataset.head())
```

```python
dataset.groupby('Category').describe()
```

```python
dataset['selected'] = dataset['Category'].apply(lambda x: 1 if x == 'Data Science' else 0)
```

```python
dataset.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(dataset.Text,dataset.selected,test_size=0.2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
v = CountVectorizer()
```

```python
X_train_count = v.fit_transform(X_train.values)
```

```python
X_train_count.toarray()[:2]
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
```

```python
model.fit(X_train_count,y_train)
```

```python
resumes = ['''Python SQL Machine Learning Deep Learning NLP Data Analysis''','''HTML CSS JavaScript React Web Design Bootstrap''']
```

```python
resumes_count = v.transform(resumes)
```

```python
model.predict(resumes_count)
```

```python
X_test_count = v.transform(X_test)
```

```python
model.score(X_test_count,y_test)
```

```python
from sklearn.pipeline import Pipeline
```

```python
clf = Pipeline([('vectorizer', CountVectorizer()),('nb', MultinomialNB())])
```

```python
clf.fit(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
clf.predict(resumes)
```
