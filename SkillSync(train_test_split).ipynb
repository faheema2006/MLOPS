```python
import pandas as pd
dataset = pd.read_csv('/content/clean_resume_dataset.csv')
print(dataset.head())
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset['Resume_Length'] = dataset['Text'].apply(len)
dataset['Skill_Count'] = dataset['Text'].apply(lambda x: len(str(x).split()))
```

```python
plt.scatter(dataset['Resume_Length'], dataset['Category'])
plt.xlabel('Resume Length')
plt.ylabel('Category')
plt.show()
```

```python
plt.scatter(dataset['Skill_Count'], dataset['Category'])
plt.xlabel('Skill Count')
plt.ylabel('Category')
plt.show()
```

```python
X = dataset[['Resume_Length', 'Skill_Count']]
dataset['Category'] = dataset['Category'].astype('category').cat.codes
y = dataset['Category']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)
```
