```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
dataset = pd.read_csv('/content/clean_resume_dataset.csv')
print(dataset.head())
```

```python
dataset['Resume_Length'] = dataset['Text'].apply(len)
dataset['Category'] = dataset['Category'].astype('category').cat.codes
```

```python
plt.scatter(dataset.Resume_Length,dataset.Category,marker='+',color='red')
plt.xlabel('Resume Length')
plt.ylabel('Category')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(dataset[['Resume_Length']],dataset.Category,train_size=0.8)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
y_predicted = model.predict(X_test)
```

```python
import math
def sigmoid(x):
    return 1 / (1 + math.exp(-x))
```

```python
def prediction_function(resume_length):
    z = 0.001 * resume_length - 2
    y = sigmoid(z)
    return y
```

```python
resume_length = 3000
prediction_function(resume_length)
```
