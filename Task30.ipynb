```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Flatten

# 1. Load and preprocess the MNIST dataset
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()

# Normalize pixel values to be between 0.0 and 1.0
X_train = X_train / 255.0
X_test = X_test / 255.0

# 2. Define the different hidden layer architectures to test
architectures = [
    [100],
    [200],
    [300],
    [100, 50],
    [200, 100],
    [300, 150],
    [512, 256],
    [512, 256, 128]
]

best_accuracy = 0
best_architecture = None

# 3. Iterate through each architecture, train, and evaluate
for arch in architectures:
    model = Sequential()

    # Flatten the 28x28 images into a 784-element vector
    model.add(Flatten(input_shape=(28, 28)))

    # Dynamically add hidden layers based on the current architecture
    for neurons in arch:
        model.add(Dense(neurons, activation='relu'))

    # Output layer for 10 classes (digits 0-9)
    model.add(Dense(10, activation='softmax'))

    # Compile the model
    model.compile(
        optimizer='adam',
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )

    # Train the model (verbose=0 keeps the output clean)
    model.fit(X_train, y_train, epochs=5, verbose=0)

    # Evaluate on the test dataset
    loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
    print(f"Architecture {arch} -> Accuracy: {accuracy:.4f}")

    # Track the highest performing configuration
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_architecture = arch

# 4. Print the final results
print("\nBest Architecture:", best_architecture)
print("Best Accuracy:", best_accuracy)
```
